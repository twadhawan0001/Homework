1. Write a function make sine wave(x, A, w) that returns:
y = A ·sin(w ·x)
where x is an array, A is amplitude, and w is angular frequency.
Create an array x using NumPy that goes from 0 to 2π with 1000 points

```python
def make_sine_wave(x, A, w):
    '''returns sine wave where x is an array, A is amplitude, and w is angular frequency'''
    y = A * np.sin(w * x)
    return y 
```

```python
x = np.linspace(0, (np.pi*2), num= 1000)
a_arr = np.array([0.5, 1, 1.5, 2, 2.5])
w_arr = np.array([1,2,3,4,5])
```

```python
matplt.figure(figsize=(6, 3))
for A, w in zip(a_arr, w_arr):
    y = make_sine_wave(x, A, w)
    matplt.plot(x, y, label=f"A={A}, w={w}")
matplt.legend()
matplt.title('Sine Waves of Different Amplitude and Frequency')
matplt.xlabel("X(0 to 2pi)")
matplt.ylabel("Sine value")
matplt.grid(True)
matplt.show()
    
```

3. Seaborn

```python
import seaborn as sns 
penguins = sns.load_dataset("penguins")
penguins.info()
```

```python
fig, axs = matplt.subplots(1,2, figsize=(10,5))
sns.scatterplot(data=penguins, ax=axs[0])
axs[0].set_title('Penguin Characteristics Quantification')
axs[0].set_xlabel('Species Number')
axs[0].set_ylabel('Value of Characteristic in (g and mm)')
axs[0].legend()
body_mass = penguins.loc[0:344, 'body_mass_g']
sns.histplot(data=penguins, x=body_mass, ax=axs[1])
axs[1].set_title('Distribution of Penguin Body mass')
axs[1].set_xlabel('Body Mass')
axs[1].set_ylabel('Count')
```
